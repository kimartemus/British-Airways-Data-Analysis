```python

#Method 2: Using VADER for Sentiment Analysis
#VADER (Valence Aware Dictionary and sEntiment Reasoner) is a lexicon and rule-based sentiment analysis tool specifically tuned for social media. 
#It provides more accurate sentiment analysis for short, informal text like tweets or reviews.

import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Load your cleaned dataset
data = pd.read_csv('/Users/artemus/Documents/My_Projects/BritishAirways/data/cleaned_BA_reviews.csv')

# Function to get sentiment using VADER
def vader_sentiment(review):
    sentiment_score = sia.polarity_scores(review)
    # Extract compound score which indicates overall sentiment
    return sentiment_score['compound']

# Apply the function to the review_text column
data['vader_sentiment'] = data['reviews'].apply(vader_sentiment)

# Display sentiment scores
print(data[['reviews', 'vader_sentiment']])

# Set up the plotting
plt.figure(figsize=(10, 6))

# Plot the distribution of the sentiment scores
sns.histplot(data['vader_sentiment'], bins=30, kde=True, color='purple')
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Number of Reviews')
plt.grid(True)

# Show the plot
plt.show()
```
